```python
import numpy as np
import pandas as pd
from scipy import stats

songs=pd.read_csv('data/tracks.csv')
songs=songs.head(25000)
```

```python
songs_for_lasso_and_ridge = songs.copy()
songs_for_lasso_and_ridge.set_index('id',inplace=True)
```

```python
def drop_columns(dataframe,to_be_deleted):
    dataframe.drop(to_be_deleted, axis=1, inplace=True)
to_be_deleted = ['id_artists', 'artists','name','release_date'] 
```

```python
songs_for_lasso_and_ridge['release_date'] = pd.to_datetime(songs_for_lasso_and_ridge['release_date'], errors='coerce')
```

```python
songs_for_lasso_and_ridge['year'] = pd.DatetimeIndex(songs_for_lasso_and_ridge['release_date']).year
songs_for_lasso_and_ridge['month'] = pd.DatetimeIndex(songs_for_lasso_and_ridge['release_date']).month
drop_columns(songs_for_lasso_and_ridge,to_be_deleted)#delete columns
```

```python
from sklearn import preprocessing

def scale_whole_df_except_valence():
    valence_only = songs_for_lasso_and_ridge['valence']
    del songs_for_lasso_and_ridge["valence"]

    x = songs_for_lasso_and_ridge #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    songs_for_lasso_and_ridge[songs_for_lasso_and_ridge.columns] = min_max_scaler.fit_transform(songs_for_lasso_and_ridge[songs_for_lasso_and_ridge.columns].to_numpy())
    songs_for_lasso_and_ridge['valence'] = valence_only
scale_whole_df_except_valence()
```

```python
songs_for_lasso_and_ridge["key"] = songs_for_lasso_and_ridge["key"].astype("category")
songs_for_lasso_and_ridge = pd.get_dummies(songs_for_lasso_and_ridge, columns=["key"])
```

```python
songs_for_lasso_and_ridge.dropna(inplace=True)
```

# AND SO IT BEGINS 

```python
songs_for_lasso_and_ridge
```

```python
x = songs_for_lasso_and_ridge['valence']
del songs_for_lasso_and_ridge["valence"]
songs_for_lasso_and_ridge['valence'] = x
```

```python
independent_variables = songs_for_lasso_and_ridge.columns.drop('valence')
y = songs_for_lasso_and_ridge['valence']
X = songs_for_lasso_and_ridge[independent_variables]
```

```python

from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(
                        X, np.ravel(y),
                test_size = 0.20, random_state = 1)
```

```python
songs_for_lasso_and_ridge
```

```python
# grid search hyperparameters for lasso regression
from numpy import arange
from pandas import read_csv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso
from sklearn import linear_model
# define model
model = Lasso()
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=8, random_state=1)
# define grid
grid = dict()
grid['alpha'] = arange(0, 1, 0.01)
grid['max_iter'] = [1000]
grid['positive'] = [False]
grid['tol'] = [0.0001,0.001,0.01]
# define search
search = GridSearchCV(model, grid, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(X_train, y_train)
# summarize
print('MAE: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
```

```python

grid_predictions = search.predict(X_test)
from sklearn import metrics 
metrics.mean_absolute_error(y_test, grid_predictions)
```

```python

# use automatically configured the ridge regression algorithm
from numpy import arange
from pandas import read_csv
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Ridge

model = Ridge()
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
grid = dict()
grid['alpha'] = arange(0, 1, 0.01)
grid['max_iter'] = [1000]
grid['positive'] = [False,True]
grid['tol'] = [0.0001,0.001,0.01]
search = GridSearchCV(model, grid, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(X_train, y_train)
# summarize
print('MAE: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
```

```python

grid_predictions = search.predict(X_test)
from sklearn import metrics 
metrics.mean_absolute_error(y_test, grid_predictions)
```
